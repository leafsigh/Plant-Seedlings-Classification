This chunk read each image files in each class files and convert them into BGR form, restore all BGR form matrix in a dictionary. The key word of the dictionary is class, and each key contains all image BGR matrices belong to this key/class

__This runs for a minute__

```python
images_per_class={}
for class_file_name in os.listdir(train_data_folder):
    class_folder_path = os.path.join(train_data_folder,class_file_name)
    class_label = class_file_name
    images_per_class[class_label]=[]
    for image_path in glob(os.path.join(class_folder_path, "*.png")):
        image_bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
        images_per_class[class_label].append(image_bgr)
```

__plt.subplots(rows,cols,figsize)__ generates an empty panel which contains rows*x*cols subplots, and figsize controls the panel's size

```python
def image_show(label):
    rows = 3
    cols = 3
    fig,axs = plt.subplots(rows,cols,figsize = (9,9))
    
    n = 0
    for i in range(0, rows):
        for j in range(0, cols):
            #axs[i, j].xaxis.set_ticklabels([])
            #axs[i, j].yaxis.set_ticklabels([])
            axs[i, j].imshow(images_per_class[label][n])
            n += 1        
```

# Processing for the Images

    changing colorspaces mean convert images from one color-space to another, like BGR to Gray, BGR to HSV etc.

For color conversion, we use the function __cv2.cvtColor(input_image, flag)__ where flag determines the type of conversion.

For BGR to Gray conversion we use the flags __cv2.COLOR_BGR2GRAY__. Similarly for BGR to HSV, we use the flag __cv2.COLOR_BGR2HSV__. To get other flags, just run following commands in your Python terminal :

```python
flags = [i for i in dir(cv2) if i.startswith('COLOR_')]
#print (flags)
```

```python
def create_mask_for_plant(image):
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    sensitivity = 30
    lower_hsv = np.array([60 - sensitivity, 100, 50])
    upper_hsv = np.array([60 + sensitivity, 255, 255])

    mask = cv2.inRange(image_hsv, lower_hsv, upper_hsv)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (11,11))
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    
    return mask
```

The next chunk helps to explore the changing between each step of __create_mask_for_plant__

```python
image0 = images_per_class['Sugar beet'][1]
image1 = cv2.cvtColor(image0, cv2.COLOR_BGR2HSV)

sensitivity = 30
lower_hsv = np.array([60 - sensitivity, 100, 50])
upper_hsv = np.array([60 + sensitivity, 255, 255])
image2 = cv2.inRange(image1,lower_hsv,upper_hsv)

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20,20))
image3 = cv2.morphologyEx(image2, cv2.MORPH_CLOSE, kernel)

fig, axs = plt.subplots(1, 4, figsize=(15, 15))
axs[0].imshow(image0)
axs[1].imshow(image1)
axs[2].imshow(image2)
axs[3].imshow(image3)
#axs[1].imshow(image_mask)
#axs[2].imshow(image_segmented)
#axs[3].imshow(image_sharpen)
```
